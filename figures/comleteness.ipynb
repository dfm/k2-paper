```python
import fitsio
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
from itertools import product
from matplotlib.ticker import ScalarFormatter
```

Load the simulations and results.

```python
injections = pd.read_hdf("data/injections.h5", "injections")
injections["inj_id"] = np.arange(len(injections))
features = pd.read_hdf("data/tces.h5", "features")
```

Apply the signal to noise cuts and see which injections are recovered.

```python
features["order"] = (features.phic_norm - features.median_phic_norm) / features.rrms_phic_norm
m = features.order > 15.0
m &= features.period > 4.0
tces = features[m]

joined = pd.merge(injections, tces[["directory", "period", "t0", "depth"]], on="directory", how="left",
                  suffixes=("", "_rec"))
joined["passed"] = (np.abs(joined.period - joined.period_rec) < 0.1) & (np.abs(joined.t0 - joined.t0_rec) < 0.1)

injections["passed"] = joined.groupby("inj_id")["passed"].sum() > 0
```

```python
m = joined.passed
x = 1e3 * joined[m].radius ** 2
pl.loglog(x, joined[m].depth, ".k", alpha=0.5)
pl.plot([0.1, 100], [0.1, 100], "g", lw=2)
pl.xlim(x.min(), x.max())
pl.ylim(0.2, 80)
pl.xlabel("injected depth")
pl.ylabel("recovered depth")
ax = pl.gca()
ax.set_xticks([1, 10])
ax.get_xaxis().set_major_formatter(ScalarFormatter())
ax.set_yticks([1, 10])
ax.get_yaxis().set_major_formatter(ScalarFormatter()) 
```

```python
candidates = fitsio.read("data/candidates.fits")
```

```python
fig, ax = pl.subplots(1, 1, figsize=(5, 5))

x = candidates["comp_period"]
lny = candidates["comp_ln_ror"] + np.log(100)
y = np.exp(lny)
yerr_m = y - np.exp(lny - candidates["comp_ln_ror_minus"])
yerr_p = np.exp(lny + candidates["comp_ln_ror_plus"]) - y

ax.errorbar(x, y, yerr=(yerr_m, yerr_p), fmt=".k", capsize=0)

ax.set_xlim(4, 70)
ax.set_ylim(1.1, 70)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("period [days]")
ax.set_ylabel("$R_\mathrm{P} / R_\star$ [\%]")
ax.set_xticks([5, 10, 20, 50])
ax.get_xaxis().set_major_formatter(ScalarFormatter()) 
ax.set_yticks([2, 5, 10, 20, 50])
ax.get_yaxis().set_major_formatter(ScalarFormatter()) 

fig.subplots_adjust(wspace=0.01, bottom=0.15, top=0.98, right=0.98, left=0.15)
pl.savefig("candidates.pdf");
```

```python
fig, ax = pl.subplots(1, 1, figsize=(5, 5))

# First: period vs. depth:
m = injections.period > 4.0
x, y = np.log10(injections[m].period), np.log10(injections[m].radius) + 2
n_all, bx, by = np.histogram2d(x, y, bins=(10, 11))

m = injections.passed
x, y = np.log10(injections[m].period), np.log10(injections[m].radius) + 2
n_rec, _, _ = np.histogram2d(x, y, bins=(bx, by))
comp = n_rec / n_all
X, Y = np.meshgrid(bx, by, indexing="ij")

cax = ax.pcolor(10**X, 10**Y, 100*comp, cmap="BuGn", vmin=0, vmax=100)

# Label the bins with their completeness percentages.
for i, j in product(range(len(bx) - 1), range(len(by) - 1)):
    x = 0.5 * (bx[i] + bx[i+1])
    y = 0.5 * (by[j] + by[j+1])
    ax.annotate("{0:.0f}".format(100 * comp[i, j]), (10 ** x, 10 ** y),
                ha="center", va="center", alpha=1.0, fontsize=10)

ax.set_xlim(np.min(10**X), np.max(10**X))
ax.set_ylim(np.min(10**Y), np.max(10**Y))
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("period [days]")
ax.set_ylabel("$R_\mathrm{P} / R_\star$ [\%]")
ax.set_xticks([5, 10, 20, 50])
ax.get_xaxis().set_major_formatter(ScalarFormatter()) 
ax.set_yticks([2, 5, 10])
ax.get_yaxis().set_major_formatter(ScalarFormatter()) 

fig.subplots_adjust(wspace=0.01, bottom=0.15, top=0.98, right=0.98, left=0.15)
pl.savefig("completeness.pdf");
```

```python
fig, ax = pl.subplots(1, 1, figsize=(5, 3.5))

# Then with magnitude:
x = injections.kepmag
m = injections.period > 4.0
n_all, bins = np.histogram(x[m], bins=12)
m = injections.passed
n_rec, _ = np.histogram(x[m], bins)
comp = n_rec / n_all
comp[~np.isfinite(comp)] = 0.0

x = np.array(zip(bins[:-1], bins[1:])).flatten()
y = 100 * np.array(zip(comp, comp)).flatten()
ax.plot(x, y, "k")
# ax.errorbar(0.5 * (bins[1:] + bins[:-1]), 100 * comp, yerr=np.sqrt(100 * comp), fmt="+k")

ax.set_xlim(np.min(x), np.max(x))
ax.set_ylim(0, 68)
ax.set_xlabel("\emph{Kepler} mag. from EPIC")
ax.set_ylabel("det. eff. [\%]")

# ax.set_yticks([1, 3, 10, 30])
# ax.get_yaxis().set_major_formatter(ScalarFormatter()) 
fig.subplots_adjust(wspace=0.01, bottom=0.2, top=0.98, right=0.96, left=0.15)
pl.savefig("completeness-mag.pdf");
```
