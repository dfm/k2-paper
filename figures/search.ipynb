```python
from __future__ import division, print_function

%matplotlib inline
%config InlineBackend.figure_format = "retina"

from matplotlib import rcParams
rcParams["savefig.dpi"] = 100
rcParams["font.size"] = 16
```

```python
import os
import numpy as np
import matplotlib.pyplot as pl
from matplotlib.ticker import MaxNLocator
```

```python
import turnstile
```

```python
lc_dir = "../../k2/lightcurves/"
epicid = "201374602"
fn = lc_dir + "c1/{0}00000/{1}000/ktwo{2}-c01_lpd-lc.fits".format(epicid[:4], epicid[4:6], epicid)
basis = lc_dir + "c1-basis.h5"
```

```python
pipe = turnstile.K2Data(basepath="./cache")
pipe = turnstile.K2Inject(pipe)
pipe = turnstile.K2Likelihood(pipe)
pipe = turnstile.OneDSearch(pipe)
pipe = turnstile.TwoDSearch(pipe)
pipe = turnstile.PeakDetect(pipe)
```

```python
q = dict(
    light_curve_file=fn,
    basis_file=basis,
    injections=[dict(
        period=18.0, t0=12.4, radius=0.02,
    )],
    time_spacing=0.01,
    durations=[0.1],
    min_period=10.,
    max_period=70.,
    alpha=np.log(3500.)-np.log(2*np.pi),
)
```

```python
r = pipe.query(**q)
```

```python
lc = r.model_light_curves[0]
```

```python
mu = lc.predict()
```

```python
pl.plot(lc.time, lc.flux-mu, ".k")
```

```python
fig = pl.figure(figsize=(6, 3))
ax = pl.axes([0.15, 0.2, 0.83, 0.78])

t = np.arange(r.min_time_1d, r.max_time_1d, r.time_spacing)

p, t0 = q["injections"][0]["period"], q["injections"][0]["t0"]
while t0 < max(t):
    ax.axvline(t0, lw=3, color="g", alpha=0.3)
    t0 += p

mu, std = r.depth_1d[:, 0], 1.0 / np.sqrt(r.depth_ivar_1d)[:, 0]
ax.plot(t, mu, "k", lw=0.7)
    
ax.set_xlim(min(t), max(t))
ax.set_ylim(-0.32, 0.65)
ax.set_ylabel("ML depth [ppt]")
ax.set_xlabel("time [BJD - 2456808]")
fig.savefig("linear.pdf")
```

```python
fig = pl.figure(figsize=(6, 3))
ax = pl.axes([0.15, 0.2, 0.83, 0.78])

p = q["injections"][0]["period"]
ax.axvline(p, lw=3, color="g", alpha=0.3)

ax.plot(r.period_2d, r.phic_same, "k", lw=0.7)

args = dict(xycoords="data", xytext=(0, 15), textcoords="offset points",
            ha="center", va="bottom", fontsize=12,
            arrowprops=dict(arrowstyle="-", shrinkA=0.0, shrinkB=6))
    
for peak, nm in zip(r.peaks[:2], ["first", "second"]):    
    ax.annotate(nm, xy=(peak["period"], peak["phic_same"]), **args)
    
ax.set_ylim(0, 110)
ax.set_xlim(11, 69)
ax.yaxis.set_major_locator(MaxNLocator(5))
ax.set_xlabel("period [days]")
ax.set_ylabel("objective")
fig.savefig("periodic.pdf")
```

```python
fig = pl.figure(figsize=(6, 3))
ax = pl.axes([0.15, 0.2, 0.83, 0.78])

p = q["injections"][0]["period"]
ax.axvline(p, lw=3, color="g", alpha=0.3)

ax.plot(r.period_2d, r.depth_2d * np.sqrt(r.depth_ivar_2d), "k", lw=0.7)

args = dict(xycoords="data", xytext=(0, 15), textcoords="offset points",
            ha="center", va="bottom", fontsize=12,
            arrowprops=dict(arrowstyle="-", shrinkA=0.0, shrinkB=6))

for peak, nm in zip(r.peaks[:2], ["first", "second"]):    
    ax.annotate(nm, xy=(peak["period"], peak["depth_s2n"]), **args)

ax.set_ylim(0, 14.8)
ax.set_xlim(11, 69)
ax.yaxis.set_major_locator(MaxNLocator(6))
ax.set_xlabel("period [days]")
ax.set_ylabel("depth S/N")
fig.savefig("depth_s2n.pdf")
```
