```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time

from vespa.transitsignal import TransitSignal
from vespa.populations import HEBPopulation, EBPopulation, BEBPopulation, PlanetPopulation

from vespa.transit_basic import MAInterpolationFunction

import logging
rootLogger = logging.getLogger()
rootLogger.setLevel(logging.INFO)
```

```python
period = 32.988
mags = {'B':15.005,  'Berr':0.06,  'V':13.496,   'Verr':0.05,   
        'g':14.223,     'gerr':0.05,  'r':12.858,  'rerr':0.04,   
        'i':11.661,  'ierr':0.08,     'J':9.763,  'Jerr':0.03,    
        'H':9.135,  'Herr':0.03,   'K':8.899,  'Kerr':0.02,   
        'W1':8.769,   'W1err':0.023,   'W2':8.668,  'W2err':0.020,    
        'W3':8.552,   'W3err':0.025,   'W4':8.587,   'W4err':0.379,
        'Kepler':12.473}

ra, dec = '11:30:14.510','+07:35:18.21'

radius = (0.394, 0.038)
mass = (0.41, 0.04) #just a guess
feh = (0.09, 0.09)
age = (9.7,0.1)
Teff = 3503
logg = 4.86

rprs = 0.0534
```

```python
lc = pd.read_csv('k2oi.csv')
lc['flux'] *= 0.001
lc['flux'] += 1
lc['ferr'] *= 0.001
close = np.absolute(lc['tst']) < 0.4
lc = lc[close]
trsig = TransitSignal(lc['tst'].values, 
                      lc['flux'].values, 
                      lc['ferr'].values, P=period,
                      name='K2OI-awesome')
trsig.MCMC(savedir='K2OI-awesome_chains')
trsig.plot(plot_trap=True)
```

```python
trsig.slopefit
```

Hmm... fit looks to be decently V-shaped ($T/\tau\sim 2$), but not actually a good fit; probably this one outlier point is hurting us:

```python
lc[(np.absolute(lc['tst']) < 0.05) & (lc['flux']>1)]
```

OK, just for illustration, let's take that point out and see what the slope looks like:

```python
newlc = lc.drop(2601)
trsig = TransitSignal(newlc['tst'].values, 
                      newlc['flux'].values, 
                      newlc['ferr'].values, P=period,
                      name='K2OI-awesome')
trsig.MCMC(savedir='K2OI-awesome_chains_refit')
trsig.plot(plot_trap=True)
```

```python
trsig.slopefit
```

OK, this honestly looks like a better fit, so we'll keep it for now.

```python
from vespa.populations import PopulationSet
popset = PopulationSet('k2oi-awesome_pops.h5')
```

```python
from vespa.fpp import FPPCalculation
import astropy.units as u
fpp = FPPCalculation(trsig, popset)
fpp['beb'].set_maxrad(8*u.arcsec) #let's be generous
```

```python
fpp.change_prior(fp_specific=0.3)
fpp.apply_secthresh(0.001) #no secondary eclipse at the 1ppt level
```

```python
fpp.FPP(verbose=True)
```

```python
fpp['eb'].lhoodplot(trsig)
fpp['beb'].lhoodplot(trsig)
fpp['heb'].lhoodplot(trsig)
fpp['pl'].lhoodplot(trsig)
```

```python
fpp.FPPsummary()
```

This planet gets the Morton validation stamp! (as long as we get rid of that outlier photometric point...)
